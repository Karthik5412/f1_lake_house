```python
import pyspark.sql.functions as f
```

```python
df = spark.read.table('f1_racing.bronze.brz_drivers')

display(df.limit(4))
```

```python
df = df.drop('number','url','source_fie','ingested_at')

display(df)
```

```python
df = df.withColumn('code', f.when(f.col("code") == r'\N', f.upper(f.substring(f.col("driverRef"),0,3)))
              .otherwise(f.col('code')))

display(df.limit(4))
```

```python
df = df.withColumnsRenamed({ 
                       'driverId' : 'driver_id',
                       'driverRef' : 'driver_ref',
                       'forename' : 'first_name',
                       'surname' : 'last_name'
})

df.show()
```

```python
df = df.withColumn('driver_ref',f.regexp_replace(f.col("driver_ref"),"_",' '))

df.show()
```

```python
df.write.format('delta')\
    .mode('overwrite')\
    .option('mergeSchema',True)\
    .saveAsTable('f1_racing.silver.slv_drivers')
```
