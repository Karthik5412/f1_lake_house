```sql
CREATE OR REPLACE VIEW f1_racing.gold.v_gold_race_analytics
AS
SELECT 
    res.result_id,
    res.grid AS starting_grid,
    res.position_text AS finishing_position_text,
    res.points AS points_scored,
    res.laps AS laps_completed,
    res.time AS race_duration,
    res.fastest_lap AS fastest_lap_number,
    res.fastest_lap_time,
    res.fastest_lap_speed,
    res.status AS race_status,
    r.raceId AS race_id,
    r.round AS race_round,
    r.date_time AS race_date,
    r.name AS race_name,
    r.winner_id,
    r.runner_id,
    d.driver_id,
    v.circuitId AS circuit_id
FROM f1_racing.gold.gold_results res
INNER JOIN f1_racing.gold.gold_races r 
    ON res.race_id = r.raceId
INNER JOIN f1_racing.gold.gold_drivers d 
    ON res.driver_id = d.driver_id
INNER JOIN f1_racing.gold.gold_circuits v 
    ON r.circuitId = v.circuitId;
```
