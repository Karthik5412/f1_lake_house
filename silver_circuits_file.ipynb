```python
import pyspark.sql.functions as F
```

```python
df = spark.read.table('f1_racing.bronze.brz_circuits')

display(df.limit(4))
```

```python
display(df.select("circuitRef").distinct())
```

```python
df = df.withColumn("circuitRef", F.initcap(F.col('circuitRef')))\
    .withColumnsRenamed({'circuitId':'circuit_id',
                         'circuitRef' : 'circuit_ref',
                         'lat' : 'latitude',
                         'lng' : 'longitude',
                         'alt' : 'altitude'})\
    .drop('source_fie','url','ingested_at')
    

display(df.limit(4))
```

```python
df = df.withColumn('latitude', F.col("latitude").cast('string'))\
    .withColumn('longitude', F.col('longitude').cast('string'))

display(df.limit(4))
```

```python
df.write.format('delta')\
    .mode('overwrite')\
    .option('mergeSchema',True)\
    .saveAsTable('f1_racing.silver.slv_circuits')
```
