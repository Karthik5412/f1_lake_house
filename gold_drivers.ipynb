```python
%sql
select 
driver_id,
count(race_id) as races
from f1_racing.silver.slv_results
group by driver_id
order by races desc
```

```python
%sql
select
*
from(
count(res.race_id) as races,
sum(res.points) as points,
min(res.fastest_lap) as fastest_lap
from f1_racing.silver.slv_drivers as dir
left join f1_racing.silver.slv_results as res
on dir.driver_id == res.driver_id
group by dir.driver_id
)
```
