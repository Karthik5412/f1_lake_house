```python
%sql
select 
driver_id,
count(race_id) as races
from f1_racing.silver.slv_results
group by driver_id
order by races desc
```

```python
%sql
select
res.driver_id,
dir.driver_ref,
dir.code,
dir.first_name,
dir.last_name,
dir.dob,
dir.nationality,
count(res.race_id) as races,
sum(res.points) as points,
min(res.fastest_lap) as fastest_lap_of_all_time,
min(res.fastest_lap_time) as fastest_lap_time_of_all_time,
max(res.fastest_lap_speed) as all_time_top_speed
from f1_racing.silver.slv_drivers as dir 
left join f1_racing.silver.slv_results as res
on dir.driver_id = res.driver_id
where res.fastest_lap != -1 or res.fastest_lap_time != -1.000 or  res.fastest_lap_speed != -1.000
group by res.driver_id,dir.driver_ref,dir.code,dir.first_name,
dir.last_name,
dir.dob,
dir.nationality
order by res.driver_id asc

```
