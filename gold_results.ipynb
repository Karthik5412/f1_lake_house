```python
%sql 
CREATE OR REPLACE TABLE f1_racing.gold.gold_results
USING DELTA
AS

SELECT
    res.result_id,
    res.race_id,
    res.driver_id,
    res.grid,
    res.position_text,
    res.points,
    res.laps,
    res.time,
    res.fastest_lap,
    res.fastest_lap_time,
    res.fastest_lap_speed,
    s.status
FROM f1_racing.silver.slv_results as res
LEFT JOIN f1_racing.silver.slv_status as s
ON res.status_id = s.statusId
```
